```r
x<-c(0,1,2,3,4,5,6)
y<-c(34000,22791,15277,10241,6865,4600,3084)
```

```r
plot(x,y,col="blue", main="Photocopier Salvage Value",xlab="# of Years Since Date of Purchase", ylab="Salvage Value (Dollars)")
```

```r
relation_nl<-nls(y~I(34000*exp(-k*x)), start=list(k=0.4))
print(relation_nl)
summary(relation_nl)
```

```r
x_seq <- seq(from=0,to=6, by=0.1)
new_relation_nl<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation_nl[i]<-predict(relation_nl, x_new)
}
plot(x,y,col="blue", main="Nonlinear Regression Example",lines(x_seq, new_relation_nl),xlab="x",ylab="y")
```
